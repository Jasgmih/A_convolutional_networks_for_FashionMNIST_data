```python
current_file_path = os.path.join(os.getcwd(),name)
data_dir = os.path.join(os.path.dirname(current_file_path),'data/FashinMNIST')
```

```python
# Download the FashionMNIST dataset to './data/FashionMNIST'

train_set = torchvision.datasets.FashionMNIST(
    root = data_dir, 
    train = True,
    download = True,
    transform = transforms.Compose([
        transforms.ToTensor(),
    ])
)
```

```python
train_set.data.shape # Containing 60000 examples, each grayscale image is in the size of [28,28]
```

```python
train_set.targets.shape
train_set.targets.max(), train_set.targets.min() # Containing 60000 corresponding labels in the range of [0,9] 
```

```python
train_set.class_to_idx # The categoties related to each integer lable
```

```python
# Visualize the images
images = train_set.data[0:40,:,:].unsqueeze(1) # take the first 40 images in the train_set, and resize to the shape of [batch_size,C,H,W]
grid = torchvision.utils.make_grid(images, nrow=10) 

plt.figure(figsize=(12,5))
plt.imshow(np.transpose(grid,(1,2,0))) 
plt.axis('off')
```
